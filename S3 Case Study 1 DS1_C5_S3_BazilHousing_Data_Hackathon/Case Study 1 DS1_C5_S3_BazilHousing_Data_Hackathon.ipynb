```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings("ignore")
```

### Perform descriptive analysis of each numerical column & categorical  column by defining a function to print the information: the number of categories present in a column, the mode of a column

```python
def num(col):
    print("mean of ",col," is ",df_brazilhouse_num[col].mean())
    print("median of ",col," is ",df_brazilhouse_num[col].median())
    print("mode of ",col," is ",df_brazilhouse_num[col].mode()[0])
```

```python
num('area')
```

```python
num('rooms')
```

```python
num('bathroom')
```

```python
num('total (R$)')
```

```python
def cat(col):
    print("Mode of ",col," is ",df_brazilhouse_cat[col].mode()[0])
    print("Unique value is ",col," are ",df_brazilhouse_cat[col].unique())
```

```python
cat('city')
```

```python
cat('animal')
```

```python
cat('furniture')
```

## LEVEL 1 ANALYSIS –UNIVARATE ANALYSIS

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
city_count=df_brazilhouse['city'].value_counts().sort_values(ascending=False)
sb.barplot(x=city_count.values,y=city_count.index)
plt.ylabel('City')
plt.xlabel('City Count')
plt.title('Number of city')

plt.subplot(122)
plt.pie(city_count,labels=city_count.index,autopct='%.2f%%',explode=[0.07,0.05,0.05,0.05,0.05])
plt.show()
```

# 1.Bachelors
Criteria:
Room<=2 
bathroom<=2 
parking spaces<=1 
furniture== furnished

```python
df_bachlor=df_brazilhouse[(df_brazilhouse['rooms']<=2) & (df_brazilhouse['bathroom']<=2) & (df_brazilhouse['parking spaces']<=1) & (df_brazilhouse['total (R$)']<=7000) & (df_brazilhouse['furniture']=='furnished')]
df_bachlor
```

## LEVEL 1 ANALYSIS –UNIVARATE ANALYSIS - Bachelor

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(df_bachlor['total (R$)'],ec='red',bins=5)
plt.xlabel('Total R$.')
plt.ylabel('count')

plt.subplot(122)
df_bachlor['total (R$)'].plot(kind='box',showmeans=True,vert=0)
plt.show()
```

```python
df_bachlor.hist(column='total (R$)',by='city',figsize=(12,8),rwidth=0.9,legend=True,xrot=45)
```

```python
c1=df_bachlor['city'].value_counts()
plt.pie(c1,labels=c1.index,autopct='%.2f%%')
plt.show()
```

<b>Interpretation: Sao Paulo, Rio de Janeiro and Porto Alegre cities have more houses are availble with furnished under R$ 7000 for bachlors

## LEVEL 2 ANALYSIS – BIVARATE ANALYSIS

```python
c1=df_bachlor[df_bachlor['city'].isin(['São Paulo','Rio de Janeiro','Porto Alegre'])]
plt.figure(figsize=(12,4))
sb.countplot(x='city',hue='animal',data=c1)
plt.show()
```

```python
per = c1['animal'].value_counts()
plt.pie(per,labels=per.index,autopct="%.2f%%")
plt.show()
```

```python
plt.figure(figsize=(12,4))
sb.lineplot(x='floor',y='total (R$)',data=c1)
```

#### Interpretation: If floor counts are increase than house price also increase

```python
sb.kdeplot(c1['total (R$)'],shade=True,color='green')
```

```python
#relation between floor and house amount
d1=round(c1['floor'].corr(c1['total (R$)']),2)
d1
```

```python
plt.xlabel('total (R$)')
plt.ylabel('floor')
plt.title('Relationship b/w floor vs Total cost(R$)')
plt.scatter(c1['total (R$)'],c1['floor'],label='r='+str(d1))  #higher milage lesser than cyl
plt.legend()
```

## LEVEL 3 ANALYSIS -- MULTIVARATE ANALYSIS

```python
cor=c1.corr()
```

```python
cor
```

```python
plt.figure(figsize=(12,6))
sb.heatmap(cor,annot=True,fmt='.2f')
```

```python
sb.pairplot(c1,hue='animal',vars=['area','rooms'],kind='reg')
plt.show()
```

```python
sb.pairplot(c1,hue='animal',kind='reg')
plt.show()
```

```python
for city in c1['city'].unique():
    c11=c1[c1['city']==city]
    plt.figure(figsize=(15,8))
    
    plt.subplot(221)
    plt.title(city)
    sb.histplot(x=c11['total (R$)'],hue=c11['animal'],kde=True,hue_order=['acept','not acept'])
    
    plt.subplot(222)
    c11['total (R$)'].plot(kind='box',title=city,showmeans=True)
plt.show()
```

# Final Interpretation:Sao Paulo and Rio de Janeiro cities have more houses are available with furnished under R$ 7000 and animal also allowed for bachlors
