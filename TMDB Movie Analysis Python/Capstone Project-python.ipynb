```python
row,col=df.shape
print("rows : ",row)
print("columns : ",col)
```

```python
import json
```

```python
def safe_json_parse(json_str):
    try:
        return json.loads(json_str.replace("'", "\""))
    except json.JSONDecodeError:
        return []

# Convert the 'production_companies' column values to strings
df['production_companies'] = df['production_companies'].astype(str)

# Create a new column 'production_companies1' by joining 'production_companies' values with commas
df['production_companies1'] = df['production_companies'].apply(lambda x: ', '.join([company['name'] for company in safe_json_parse(x)]))

# Display the DataFrame with the new 'production_companies1' column
print(df[['original_title', 'production_companies1']])
```

```python
# Create a new column 'Genres' with concatenated genre names
df['Genres1'] = df['genres'].apply(lambda x: ', '.join([genre['name'] for genre in json.loads(x.replace("'", "\""))]))

# Display the DataFrame with the 'Genres' column
print(df[['original_title', 'Genres1']])
```

<b>Task 7: How are popularities of movies related with the movie budgets?Are they correlated or totally uncorrelated with each other?write the interpretation of your analysis.

```python
d1=round(new_df['budget'].corr(new_df['popularity']),2)
d1
```

```python
plt.xlabel("budget")
plt.ylabel("popularity")
plt.title("co-relation between budget and popularity")
plt.scatter(new_df['budget'],new_df['popularity'],label='r='+str(d1))
plt.legend()
plt.show()
```

<b>Interpretation: Movie budgets and popularity are correlated, but not perfectly.there is a moderate positive correlation (0.43) between movie budgets and movie popularity. This suggests that, on average, movies with higher budgets tend to be more popular. However, other factors also play a significant role in a movie's success, and budget alone does not guarantee popularity.

<b>Task 10:sort the data in descending order based on revenue and filter the top 500 movies. find the measures of central tendancy for the following columns using the filter data.1.budget, revenue,runtime and perform outlier analysis for above 3 columns using box plots.

```python
df_rev=new_df.sort_values(by=['revenue'], ascending=False)[:500]
df_rev
```

```python
cen_ten=pd.DataFrame()
cen_ten['central tendancy']=['Mean','Median','Mode']
cen_ten
```

```python
bud=[]
bud.append(df_rev['budget'].mean())
bud.append(df_rev['budget'].median())
bud.append(df_rev['budget'].mode()[0])
cen_ten['Budget']=bud
cen_ten
```

```python
rev=[]
rev.append(df_rev['revenue'].mean())
rev.append(df_rev['revenue'].median())
rev.append(df_rev['revenue'].mode()[0])
cen_ten['revenue']=rev
cen_ten
```

```python
run=[]
run.append(df_rev['runtime'].mean())
run.append(df_rev['runtime'].median())
run.append(df_rev['runtime'].mode()[0])
cen_ten['runtime']=run
cen_ten
```

```python
out=df_rev
out
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.boxplot(out['budget'],showmeans=True, vert=1)
plt.title('Budget Box Plot')

plt.subplot(1, 3, 2)
plt.boxplot(out['revenue'],showmeans=True, vert=1)
plt.title('Revenue Box Plot')

plt.subplot(1, 3, 3)
plt.boxplot(out['runtime'],showmeans=True, vert=1)
plt.title('Runtime Box Plot')

plt.tight_layout()
plt.show()
```
